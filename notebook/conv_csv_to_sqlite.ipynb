```python
import sqlite3
import pandas as pd

# File paths
csv_file = r"C:\Users\hamen\OneDrive\Desktop\project\dataset\update_cust.csv"  # Your CSV file
sqlite_db = r"C:\Users\hamen\OneDrive\Desktop\project\database\project1.db"     # SQLite database file
table_name = "customers"       # Table name

# Load CSV into pandas DataFrame
df = pd.read_csv(csv_file)

# Connect to SQLite
conn = sqlite3.connect(sqlite_db)

# Write DataFrame to SQLite (auto-creates schema from CSV headers)
df.to_sql(table_name, conn, if_exists="replace", index=False)

# Commit and close
conn.commit()
conn.close()

print(f"CSV data from {csv_file} successfully imported into {sqlite_db} as table '{table_name}'!")
```
